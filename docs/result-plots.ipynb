```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(run: np.ndarray, frames: list):
    fig, axs = plt.subplots(ncols=len(frames), figsize=(1+len(frames)*4, 4))
    for i, f in enumerate(frames):
        axs[i].imshow(run[f], origin="lower", cmap="inferno")
        axs[i].set_xticks([])
        axs[i].set_yticks([])
    fig.tight_layout()
    return fig
```

```python
run = np.load("../../simulations/morpheus-singlecell/results-one/run_0.npy")
run.shape
```

```python
plot(run, [0,12,17])
```
